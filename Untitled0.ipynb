```python
import matplotlib.pyplot as plt
import numpy as np
import math

x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine wave")
plt.show()
```

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(2,3)

import numpy as np
x= np.arange(1,5,0.2)
axes[0][0].plot(x,x*x)
axes[0][0].set_title('square')

axes[0][1].plot(x,np.sqrt(x))
axes[0][1].set_title('square root')

axes[0][2].plot(x,np.exp(x))
axes[0][2].set_title('exp')

axes[1][0].plot(x,np.log10(x))
axes[1][0].set_title('log')

axes[1][1].plot(x,2*x+5)
axes[1][1].set_title('Linear')

axes[1][2].plot(x,2*x**2-11*x+10)
axes[1][2].set_title('Polynomial')

plt.tight_layout() #spaces out the plots
plt.show()#display all the plot
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
image =img.imread("flowers.jpg")
fig,axes=plt.subplots(2,4,figsize=(10,7))

print(image)
print()

axes[0][0].imshow(image)
axes[0][1].imshow(image[:,:,0],cmap="Reds")
#Show only the Red channel of the image using red shades
axes[0][2].imshow(image[:,:,1],cmap="Greens")
axes[0][3].imshow(image[:,:,2],cmap="Blues")

axes[1][0].imshow(image,cmap="viridis")
axes[1][1].imshow(image[:,:,0],cmap="gray") # red channel in gray color
axes[1][2].imshow(image[:,:,1],cmap="gray") #green channel in gray color
axes[1][3].imshow(image[:,:,2],cmap="gray") #blue channel in gray color

plt.tight_layout() # space out the plots
plt.show()#display all the plots
```
